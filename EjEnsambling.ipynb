```python
del df['Name']
del df['PassengerId']
```

```python
#VARIABLE CABIN

inputer = SimpleImputer(missing_values = np.nan, strategy='constant', fill_value='')
df['Cabin'] = inputer.fit_transform(df[['Cabin']])
df['Cabin']
```

```python
df['Cabin'] = [i[0] if i != '' else '' for i in df['Cabin'] ]
df
```

```python
df['Cabin'].replace('', 'NoCabin', inplace=True)
df
```

```python
df.groupby(['Cabin','Survived']).size()
df['Cabin'].replace(['A','F','G','T'], 'AFGT', inplace=True)
df.groupby(['Cabin','Survived']).size()
```

```python
# varbiale ticket

df['Ticket'] = [i.lower() for i in df['Ticket']]
df
```

```python
df['Ticket'] = [i.split()[0] if len(i.split()) > 1 else '' for i in df['Ticket']]
df
```

```python
lista_valores_malos = ['1','2','3','4','5','6','7','8','9','/','.']
lista_total = []
for i in df['Ticket']:
    for j in lista_valores_malos:
         i = i.replace(j,'')
    lista_total.append(i.strip(' '))
df['Ticket'] = lista_total
df
```

```python
df.groupby(['Ticket','Survived']).size()
df['Ticket'].replace(['as','c','coston','fa','fc','fcc','pp','ppp','sc','sca','scah','scow','sop','sopp','sotono','sp','swpp','wep','casoton'], 'Agrupacion', inplace=True)
df.groupby(['Ticket','Survived']).size()
```

```python
#Missing values

imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
df['Age'] = imputer.fit_transform(df[['Age']])
df
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'constant', fill_value = 'S')
df['Embarked'] = imputer.fit_transform(df[['Embarked']])

df.isnull().sum()
```

```python
df.groupby(['Parch','Survived']).size()
df['Parch'].replace([3,4,5,6], 3456, inplace=True)
df.groupby(['Parch','Survived']).size()
```
